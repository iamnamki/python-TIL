```python
from bs4 import BeautifulSoup
import requests


class Games():
    def __init__(self):
        url = 'https://play.google.com/store/apps/category/GAME/collection/topselling_paid'
        response = requests.get(url)
        #html 읽음 
        html = BeautifulSoup(response.text, 'html.parser')
        #tag 가져오기 
        self.tags = html.select('div.card.no-rationale.square-cover.apps.small')
     
    games = []
    def get_list(self):
        for tag in self.tags :
            title = tag.select("a.title")
            self.title = title[0].text
            subtitle = tag.select("a.subtitle")
            self.subtitle = subtitle[0].text
            ratingtag = tag.select("div.current-rating")
            rating = ratingtag[0].get("style")
            rating = rating.split(':')
            self.rating = rating[1][ : -1]
            price = tag.select("span.display-price")
            self.price = price[0].text
            self.games.append(self.__str__())
            
            
    def write_csv(self):
        with open("games.csv", "w", encoding='utf-8') as file:
            file.write("게임명\t\t제조사\t\t가격\t\t평점\n")
            for game in self.games:
                file.write(str(game) + "\n")
    def read_csv(self):
        with open("games.csv", "r", encoding='utf-8') as file:            
            print(file.read())
                
    def __str__(self):
        return "{}\t{}\t{}\t{}".format(self.title, self.subtitle, self.price, self.rating)

g = Games()
# # url에서 필요한 데이터를 crawling 
g.get_list()
# # 데이터를 파일에 write
g.write_csv()
# # 파일에서 데이터를 read & print
g.read_csv()
```
